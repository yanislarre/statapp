```python
import pandas as pd

file_id = "11nqJ9AiQR9rf9A6lI1Jg8_QcX4e3hqYW" # téléchargement du fichier Global Debt Database
url = f"https://drive.google.com/uc?export=download&id={file_id}"

df_global_debt_OECD = pd.read_csv(url, skiprows=3)

print(df_global_debt_OECD.head())
```

```python
# on drop public sector debt et non financial public sector debt pour cette partie de l'analyse

df_global_debt_OECD2 = df_global_debt_OECD.drop(columns=['Public sector debt (percent of GDP)', 'Nonfinancial public sector debt (percent of GDP)'])
```

```python
# regardons les NaN pour chaque année 
# On groupe par 'Year', et pour chaque groupe on additionne toutes les valeurs nulles
nan_par_annee = df_global_debt_OECD2.groupby('Year').apply(lambda x: x.isnull().sum().sum())

# Affichage du résultat
print(nan_par_annee)
```

```python
# Trier les résultats par ordre décroissant
nan_par_annee_trié = nan_par_annee.sort_values(ascending=False)

# Afficher le résultat en liste
print(nan_par_annee_trié.reset_index().values.tolist())
```

```python
# regardons pour chaque année combien de pays ont au moins un Nan 

# isnull().any(axis=1) renvoie True si au moins une colonne de la ligne est vide
df_global_debt_OECD2['a_un_nan'] = df_global_debt_OECD2.isnull().any(axis=1)

# Grouper par année et compter le nombre de True
lignes_nan_par_annee = df_global_debt_OECD2.groupby('Year')['a_un_nan'].sum().astype(int)

# Afficher sous forme de liste [Année, Nombre de lignes avec NaN]
resultat_liste = lignes_nan_par_annee.reset_index().values.tolist()
print(resultat_liste)
```

Il y a 38 pays membres de l'OCDE, cette dernière cellule nous permet de regarder par année, sur combien de pays disposons nous des données. On remarque qu'avant 1995 on ne dispose des données complètes que quelques pays. À partie de 1995, on dispose des données complètes de plus de la moitié des pays de l'OCDE. 
Cela nous pousse à baser notre étude sur les années 1996-2024. 

```python
# on ne conserve que les années strictement supérieures à 1995 
df_global_debt_OECD2 = df_global_debt_OECD2[df_global_debt_OECD2['Year'] > 1995]
df_global_debt_OECD2 = df_global_debt_OECD2.drop(columns='a_un_nan')
df_global_debt_OECD2.head()
```
